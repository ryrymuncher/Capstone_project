```python
#import libraries
import pandas as pd
import xml.etree.ElementTree as et
```

```python
#filepath
bib_xml = 'D:/Users/Ryan/SMU classwork/MSDS_6120/BIBLIOGRAPHIC_12731526450003716_1.xml'
```

```python
#find root element from xml 
tree = et.parse(bib_xml)
root = tree.getroot()
```

```python
#parse title, author, and description from xml into python dictionary
item_records = []
for elm in root.findall("./record"):
    description = elm.find("./datafield[@tag='520']/subfield[@code='a']")
    if description is None:
        continue
    title = elm.find("./datafield[@tag='245']/subfield[@code='b']")
    if title is None:
        title = ""
    else:
        title = title.text
    author = elm.find("./datafield[@tag='100']/subfield[@code='a']")
    if author is None:
        author = ""
    else: 
        author = author.text
    item_records.append({"title":title,"author":author,"description":description.text})
print(item_records)
```

```python
#convert dictionary into a pandas dataframe
catalog = pd.DataFrame(item_records)
```
